```python
import pandas as pd
import numpy as np
sales=pd.DataFrame({
    "sales_id":[101,102,103,104,105,106,107,108,109,110],
    "Product_name":["pen","pencil",np.nan,"pencil","laptop","mobile","laptop","mobile","pen","pencil"],#commom -- pen,pencil, laptop,mobile
    "product_price":[10,15,10,15,np.nan,150,100,150,10,15],#pen- 10,pencil=15,laptop=100,mobile=150
    "qty":["1","2","3",np.nan,"5","6","7","8","9","10"]
}
)
sales
```

```python
# to convert datatype of column
print(sales.dtypes)

print(sales.isnull().sum())
```

```python
#sales["sales_id"]=sales["sales_id"].astype("object")
#print("------------",sales.dtypes)
# with the help of pd.to_numeric
sales["qty"]=pd.to_numeric(sales["qty"],errors="coerce")
sales.dtypes
```

```python
sales.isnull().sum()
```

```python
# to fill missing values : fillna , replace, interpolate, dropna, ffill, bfill, method, mean, median, mode, constant

sales["Product_name"]=sales["Product_name"].fillna(method="ffill")
sales.isnull().sum()
```

```python
sales.describe(include="number")
```

```python

# use method to fill missing values

sales["qty"]=sales["qty"].fillna(sales["qty"].mean())
sales["product_price"]=sales["product_price"].fillna(100)

print(sales.isnull().sum())
```

```python
sales
```

```python
# to add new column with logic
sales["total_amount"]=sales["product_price"]*sales["qty"]
```

```python
# group by function
a=sales.groupby("Product_name")["total_amount"].sum()
a=pd.DataFrame(a.reset_index()) # to convert series to dataframe and reset index to make it as column 

#if i want other aggregation function in same group by
b=sales.groupby("Product_name").agg({"total_amount":["sum","mean","max","min","count","std","var","median"]})
b=pd.DataFrame(b.reset_index())
b.columns=["Product_name","total_amount_sum","mean","max","min","count","std","var","median"]
b
```

```python
# how to make pivot table
#sales.pivot_table(index="Product_name",values="total_amount",aggfunc="sum")

# more pivot table
a=sales.pivot_table(index="Product_name",values="total_amount",aggfunc=["sum","mean","max","min","count"]).reset_index()
a
```
