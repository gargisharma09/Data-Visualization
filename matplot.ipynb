```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [15,10,20,53]

plt.plot(x, y)
plt.show
```

```python
x = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'] 
y = [10,10,8,30,15,12,15] 

plt.plot(x, y) 
plt.title('Weekly Sales')
plt.xlabel('Day of Week') 
plt.ylabel('Sales Per Day')
plt.show 
```

```python
x = [1,2,3,4] 
y = [1000,1500, 1290, 2000] 

plt.plot(x, y, color='green', linestyle='-', linewidth=1, marker='o', label='2025 Sales Data')
plt.title('Monthly Sales Data Report')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend(loc='lower right', fontsize=10)
plt.grid(color='grey', linestyle=':', linewidth=1)
plt.ylim(1000,2000)
plt.xticks([1,2,3,4], ['M1', 'M2', 'M3', 'M4'])
plt.show()
```

```python
x = [1,2,3,4] 
y = [1000,1500, 1290, 2000] 

plt.bar(x, y, color='pink', label='2025 Sales')
plt.title('Monthly Sales Data Report')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend(loc='lower right', fontsize=10)
plt.xticks([1,2,3,4], ['M1', 'M2', 'M3', 'M4'])
plt.grid(linestyle='-', linewidth=0.3)
plt.show()
```
