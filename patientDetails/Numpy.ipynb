# COMBINING/SPLITTING

```python
# concatenate
arr1=np.array([[1,2,3],
              [8,8,8]])
arr2=np.array([[4,5,6],
              [7,7,7]])
comb_arr =  np.concatenate((arr1,arr2),axis=1) # - Adds arr2 as rows to the end of arr1  --- All the input array must have same number of dimension

print(comb_arr)
```

```python
# split
np.split(arr1,2) # - Splits arr into 3 sub-arrays
```

```python
# hsplit
np.hsplit(arr1,3) #- Splits arr horizontally on the 5th index
```
